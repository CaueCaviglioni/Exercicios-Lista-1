```python
import sympy as sp

# Capítulo 7, exercício 2, letra 'e', página 320
# Calcular o produto interno entre f(x) e g(x)
# f(x) = |x - 1/2|
# g(x) = 1/2 - |x - 1/2|

x = sp.symbols('x')
# Cada variável fg corresponde as funções encontradas após a análise modular citada no caderno
fg_1 = (1/2)*(-x + 1/2)
fg_2 = (1/2)*(x - 1/2)
fg_3 = pow(x,2) - x + 1/4
# Calcula e mostra o resultado simbólico de cada uma das integrais
I1 = sp.integrate(fg_1, x)
print("I1 = ", I1)
I2 = sp.integrate(fg_2, x)
print("I2 = ", I2)
I3 = sp.integrate(fg_3, x)
print("I3 = ", I3)
# Calcula e mostra o resultado numérico das integrais
I1 = sp.integrate(fg_1, (x, 0, 1/2))
I2 = sp.integrate(fg_2, (x, 1/2, 1))
I3 = sp.integrate(fg_3, (x, 0, 1))
I = I1 + I2 + I3
print("Produto interno de f(x)*g(x) = ", I)
```
