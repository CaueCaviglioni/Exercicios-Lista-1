```python
import sympy as sp

# Capítulo 7, exercício 3, letra 'a', página 321
# Calcular o produto interno entre f(x) e g(x) com uma função peso r(x)
# f(x) = 1 - 2*x
# g(x) = e^(-x)
# r(x) = e^x

x = sp.symbols('x')
# Definindo as funções do exercício
f = 1 - 2*x
g = pow(sp.E, -x)
r = pow(sp.E, x)
# Calculando a integral simbólica
I_simb = sp.integrate(f*g*r, x)
print("Integral simbólica: ", I_simb)
# Calculando a integral numérica
I_num = sp.integrate(f*g*r, (x, 0, 1))
print("Integral numérica: ", I_num)
```
