```python
import sympy as sp

# Capítulo 7, exercício 3, letra 'e', página 328
# Calcular ||f||
# f(x) = ln(x+1)

x = sp.symbols('x')
f = sp.ln(x + 1)
# Calcula a norma do vetor f simbolicamente e numericamente
norma_f_simb = sp.sqrt(sp.integrate(pow(f,2), x))
print("Valor simbólico da norma")
print("||f|| = ", norma_f_simb)
print()
print("Substituindo os limites de integração")
norma_f_num = pow(sp.integrate(pow(f,2), (x, 0, 1)), 1/2)
print("||f|| = ", norma_f_num)
print()
print("Valor numérico da integral: ", sp.N(norma_f_num))
```
